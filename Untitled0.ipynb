```python
import pandas as pd

dataset = pd.read_csv('/content/top250-00-19.csv')
dataset.head()
```

```python
dataset
```

```python
dataset['Name'] = dataset.Name.fillna(dataset.Name.mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
dataset['Name'] = lb.fit_transform(dataset.Name)
dataset['Position'] = lb.fit_transform(dataset.Position)
dataset['Team_from'] = lb.fit_transform(dataset.Team_from)
dataset['League_from'] = lb.fit_transform(dataset.League_from)
dataset['League_to'] = lb.fit_transform(dataset.League_to)
dataset['Team_to'] = lb.fit_transform(dataset.Team_to)
dataset['Season'] = lb.fit_transform(dataset.Season)
dataset['Market_value'] = lb.fit_transform(dataset.Market_value)
```

```python
dataset
```

```python
data2 = dataset.drop(columns = ['Market_value'])
x = dataset.drop(columns = ['Transfer_fee'])
y = dataset['Transfer_fee']
```

```python
from sklearn.preprocessing import LabelEncoder

# For Name
le_name = LabelEncoder()
dataset['Name_encoded'] = le_name.fit_transform(dataset['Name'])

# For other columns
le_position = LabelEncoder()
dataset['Position'] = le_position.fit_transform(dataset['Position'])
# ... and so on for other categorical columns
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(
    x,y, random_state=42, test_size= 0.2
)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

models = [RandomForestRegressor(), DecisionTreeRegressor(), LinearRegression()]

errors = {}

for model in models:
    model.fit(xtrain,ytrain)
    name = type(model).__name__
    score = model.score(xtest,ytest)*100
    errors[name] = score
```

```python
errors
```

```python
import matplotlib.pyplot as plt

plt.bar(errors.keys(),errors.values())
```

```python
import matplotlib.pyplot as plt

# Sample data for demonstration (replace with your actual 'errors' dictionary)

# Define a cute pastel color palette
cute_colors = ['#FFB6B9', '#FFDAC1', '#E2F0CB', '#B5EAD7', '#C7CEEA']

# Create the bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(errors.keys(), errors.values(), color=cute_colors[:len(errors)])

# Add rounded edges to bars
for bar in bars:
    bar.set_linewidth(0)
    bar.set_edgecolor('none')
    bar.set_zorder(3)

# Add labels and title with cute fonts
plt.title('🐾 Chart 🐾', fontsize=16, fontweight='bold', color='#A974BF')
plt.xlabel('Models', fontsize=12, color='#6C5B7B')
plt.ylabel('R values', fontsize=12, color='#6C5B7B')

# Set background color
plt.gca().set_facecolor('#FFF0F5')  # Light lavender pink
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Remove top and right border
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
